```python
master_df_clean = master_df[ master_df['auc_score'] <= 1 ]


from scipy.stats import skew, kurtosis
skewness = skew(master_df['auc_score'])
kurt = kurtosis(master_df['auc_score'], fisher=True) 
kurt_99 = kurtosis(master_df_clean['auc_score'], fisher=True)
skew_99 = skew(master_df_clean['auc_score'])
print("Skewness:", skewness)
print("Skewness: ", skew_99)
print("Kurtosis (Fisher):", kurt)
print("Kurtosis (Fisher):", kurt_99)

master_df['auc_score'].describe()
```

```python
import statsmodels.formula.api as smf
from patsy.contrasts import Sum

# MODEL I

model = smf.mixedlm("auc_score ~ C(Classifier_name, Sum)",
                    data=master_df,
                    groups=master_df["ds_name"]).fit()
print(model.summary())
```

```python
group_variance = model.cov_re.iloc[0, 0]
print("Group variance:", group_variance)
print("STandard Deviation: ", group_variance**0.5)

model.scale

```
